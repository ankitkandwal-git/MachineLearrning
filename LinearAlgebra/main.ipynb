Maths foundation for Machine Learning : Linear Algebra & Calculus


```python
### scaler , vector , matrix , tensor
import numpy as np
scalar = 5 
vector = np.array([1,2,3])
matrix = np.array([[1,2,3],[4,5,6]])
tensor = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print("Scalar :",scalar)
print("Vector: ",vector)
print("Matrix :",matrix)
print("Tensor :",tensor.shape)
```

```python
### aritemetic operations on vectors nd matrices 
a = np.array([2,6,3])
b = np.array([1,5,4])
dot = np.dot(a,b)
cross = np.cross(a,b)
tensor1 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
tensor2 = np.array([[[9,8],[7,6]],[[5,4],[3,2]]])
print("Addition :",a+b)
print("Dot Product :", dot)
print("Cross Product :", cross)
print("Tensor Addition :",tensor1 + tensor2)
print("Tensor Multiplication :",tensor1 * tensor2)
```

```python
### transpose of matrix and tensor
matrix = np.array([[1,2,3],[4,5,6]])
tensor = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
tranpose_matrix = np.transpose(matrix)
transpose_tensor = np.transpose(tensor, (1,0,2))
print("Transpose of Matrix:\n", tranpose_matrix)
print("Transpose of Tensor:\n", transpose_tensor)
```

```python
### Determinant and Inverse and rank of matrix
matrix = np.array([[4,7],[2,6]])
print("Determinant :",np.linalg.det(matrix))
print("Inverse :", np.linalg.inv(matrix))
print("Rank :",np.linalg.matrix_rank(matrix))
```

```python
### eigen values and eigen vectors
matrix = np.array([[4, -2],[1, 1]])
eigen_values, eigen_vectors = np.linalg.eig(matrix)
print("Eigen Values :",eigen_values)
print("Eigen Vectors :\n",eigen_vectors)
```

```python
### calculus operations
import matplotlib.pyplot as plt
x  = np.linspace(0,10,100)
y = x**2 
dy_dx = np.gradient(y,x)
plt.plot(x,y,label='y=x^2')
plt.plot(x,dy_dx,label="dy/dx")
plt.legend()
plt.show()
```

```python
### partial derivatives
def f(x,y):
    return x**2 + y**2 
x,y = (2,8)
grad_x = 2*x 
grad_y = 2*y 
print("Partial Derivative w.r.t x :",grad_x)
print("Partial Derivative w.r.t y :",grad_y)    
```

```python
### chain rule 
#y = (2x+3)^2
x = 2 
inner = 2*x+3
outer = inner**2
d_outer_d_inner = 2*inner
d_inner_d_x = 2
dy_dx = d_outer_d_inner * d_inner_d_x
print("dy/dx using Chain Rule :",dy_dx)
```
